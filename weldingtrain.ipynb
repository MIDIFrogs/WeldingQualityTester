# Preparation

```python
from ultralytics import YOLO
import torch
```

```python
import os
os.environ["WANDB_DISABLED"] = "true"
```

# Select one of the following models

```python
yolo = YOLO("yolov8n.pt")
```

```python
yolo = YOLO("yolov8l.pt")
```

# Train the model

```python
torch.multiprocessing.freeze_support()
epoch = 180
yolo.train(
    data = "/kaggle/input/welding/dataset/data.yaml",
    epochs=epoch,
    time=1,
    patience=epoch//2,
    batch=0.8,
    exist_ok=True,
    imgsz=1280, 
    save_period=25, 
    cache=True, 
    lr0=0.1,
    lrf=0.0001,
    device=0, 
#     project="working/Checkpoints", 
#     name="IHopeThisIsLastOneM",
    #resume=True,
    plots=True
    )
yolo.save("./last.pt")
```
